```r
cor_mat = cor(ecommerce_customers[, 4:ncol(ecommerce_customers)])
plot_data <- melt(round(cor_mat, 2))

options(repr.plot.width = 10, repr.plot.height = 8)

ggplot(plot_data, aes(Var1, Var2, fill = value)) + 
geom_tile() +
scale_fill_distiller(palette = "Greys", direction = 1) +
geom_text(aes(label = value), size = 8) +
xlab("Column") +
ylab("Column") +
ggtitle("Columns correlation heatmap") +
theme_bw() +
theme(text = element_text(size=20), plot.title = element_text(hjust = 0.5), axis.text.x = element_text(angle = 45, hjust = 1))

options(repr.plot.width = 10, repr.plot.height = 6)
```

- features have very low correlation with each other
- estimation of relationship between individual feature and target is less error prone

```r
mod <- lm(`Yearly Amount Spent` ~ `Length of Membership`, data = ecommerce_customers)

summary(mod)
```

- model with Length of Membership as single independent variable is better than intercept only model
- for this data, per one unit increase in Length of Membership model give Yearly Amount Spent increased by 64.219
- model helped capture 65% of variance in target variable

```r
preds <- tibble(ecommerce_customers)

preds$mod1_pred = predict(mod, ecommerce_customers)
preds$mod1_res = mod$residuals

mean_mod1_res <- mean(preds$mod1_res)

ggplot(data = preds) + 
geom_point(mapping = aes(x = mod1_pred, y = mod1_res), size = 4, alpha = 0.8) +
geom_hline(yintercept = mean_mod1_res, color="red") +
xlab("Predicted") +
ylab("Residual") + 
ggtitle("Predicted Yearly Amount Spent v.s. Residual") + 
conf
```

```r
ggplot(data = preds) + 
geom_point(mapping = aes(x =`Length of Membership`, y = `Yearly Amount Spent`), size = 4, alpha = 0.8) + 
geom_line(mapping = aes(x =`Length of Membership`, y = mod1_pred), col = "red") +
xlab("Length of Membership") +
ylab("Yearly Amount Spent") + 
ggtitle("Fitted model for Yearly spending using Membership length") + 
conf
```

```r
mod2 <- lm(`Yearly Amount Spent` ~ `Length of Membership` + `Time on App`, data = ecommerce_customers)

summary(mod2)
```

- additional 23 % change in dependent variable can be captured after including Time on App as feature
- the greater Length of Membership and Time on App, the higher Yearly Amount Spent
- model cannot be used with new or no app interaction customers

```r
preds$mod2_pred <- predict(mod2, ecommerce_customers)
preds$mod2_res <- mod2$residuals

mean_mod2_res <- mean(preds$mod2_res)

ggplot(data = preds) + 
geom_point(mapping = aes(x = mod2_pred, y = mod2_res), size = 4, alpha = 0.8) +
geom_hline(yintercept = mean_mod2_res, color="red") +
xlab("Predicted") +
ylab("Residual") + 
ggtitle("Predicted Yearly Amount Spent v.s. Residual") + 
conf
```

```r
mod3 <- lm(`Yearly Amount Spent` ~ `Length of Membership` + `Time on App` + `Avg. Session Length` + `Time on Website`, data = ecommerce_customers)

summary(mod3)
```

- it's unclear to conclude how Time on Website have effect on dependent variable in population level
- model suggest that the more lengthy customer's Sessions, the higher Yearly Amount Spent is
- model cannot be used with new or no interaction customers

```r
preds$mod3_pred <- predict(mod3, ecommerce_customers)
preds$mod3_res <- mod3$residuals

mean_mod3_res <- mean(preds$mod3_res)

ggplot(data = preds) + 
geom_point(mapping = aes(x = mod3_pred, y = mod3_res), size = 4, alpha = 0.8) +
geom_hline(yintercept = mean_mod3_res, color="red") +
xlab("Predicted") +
ylab("Residual") + 
ggtitle("Predicted Yearly Amount Spent v.s. Residual") + 
conf
```

```r
shapiro.test(preds$mod1_res)
shapiro.test(preds$mod2_res)
shapiro.test(preds$mod3_res)
```

- failed to reject the null hypothesis saying residuals come from normally distributed population
